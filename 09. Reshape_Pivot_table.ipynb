# Tabela Dinâmica

Uma tabela dinâmica é uma tabela de valores agrupados que agrega os itens individuais de uma tabela mais extensa em uma ou mais categorias discretas. Esse resumo pode incluir somas, médias ou outras estatísticas, que a tabela dinâmica agrupa usando uma função de agregação escolhida aplicada aos valores agrupados

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/air_quality_long.csv')
```

```python
df
```

###### Tabela Dinâmica (Pivot Table)

```python
df
```

```python
pd.pivot_table(data=df, columns='date.utc', values='value')
```

```python
pd.pivot_table(data=df, index='date.utc')
```

```python
df.pivot(columns='location', values='value')
```

```python
pd.pivot_table(data=df, columns='location', values='value', aggfunc='mean')
```
