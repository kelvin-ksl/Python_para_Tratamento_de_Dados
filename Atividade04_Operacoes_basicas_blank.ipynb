- **Crie uma nova coluna ``london_mg_per_cubic`` para concentração de NO2 in mg/m³**

Assuma que a temperatura de 25 graus Celsius e pressão de 1013 hPa. Use o o fator de conversão é 1,882)

```python
df['london_mg_per_cubic'] = df['station_london'] * 1.882
```

- **Crie uma nova coluna ``ratio_paris_antwerp`` fazendo a relação entre ``station_paris`` e ``station_antwerp``**

**Renomeie as colunas**
- station_antwerp para BETR801
- station_paris para FR04014
- station_london para London Westminster

```python
df.rename(columns={'station_antwerp': ' BETR801', 'station_paris': 'FR04014', 'station_london': 'Westminster'}, inplace=True)
df.head(0)
```

- **Mude o nome de todas as colunas usando lower case**

```python
df.columns= df.columns.str.lower()
df.head(0)
```
