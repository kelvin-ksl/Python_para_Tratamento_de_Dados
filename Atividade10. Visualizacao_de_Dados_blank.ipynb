## Atividade de Visualização de dados usando o Pandas

- 1. Plotando gráficos de linhas
- 2. Plotando gráficos de disperssão
- 3. Plotando grádicos de barra
- 3. Plotando gráficos de boxplot
- 5. Plotando gráficos de área

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**1. Leia o dataset ``air_quality_no2`` e atribua em uma variável df**

```python
df = pd.read_csv('air_quality_no2.csv')
df
```

**2. Plotando um único gráfico de linhas para as series ``station_paris``, ``station_london``, ``sation_antwerp``**

```python
df[["station_paris","station_london","station_antwerp"]].plot()
```

**3. Plotando um único gráfico de linhas para as series ``station_paris``**

```python
df["station_paris"].plot()
```

**4. Plotando dados de barras usando as 5 primeiras linhas da coluna ``station_london``**

```python
df["station_london"].hist(grid=False, figsize=(5,4))
```

**5. Plotando gráficos de dispersão ``station_london`` e ``station_paris``**

```python
df.plot.scatter(x = "station_london", y = "station_paris", alpha = 0.5)
```

**6. Plotando gráfico de boxblot**

```python
df.plot.box()
```

**7. plotando gráficos de are e mude o figsize para (12,4)**

```python
df.plot.area(alpha = 0.5, figsize=(12,4))
```

```python
df.plot.area(figsize=(12, 4), subplots=True)
```
